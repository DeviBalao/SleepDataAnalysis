```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sleep_health = pd.read_csv("data/Sleep_health_and_lifestyle_dataset.csv")

sleep_health.plot(kind = "scatter", marker = "o", s = 45, color = "green", 
                  x = "Sleep Duration", y = "Quality of Sleep", alpha=0.5,
                  title = "Sleep Duration Vs Quality of Sleep")
plt.show()
```

```python
female_data = sleep_health[sleep_health['Gender'] == 'Female']

#female sleep duration and quality of sleep
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#Female Sleep Duration vs Stress Level
female_data.plot(ax=ax1, kind="scatter", marker="o", s=45, color="pink",
                 y="Stress Level", x="Sleep Duration", alpha=0.5,
                 title="Female Stress Level Vs Sleep Duration")

#Female Quality of Sleep vs Stress Level
female_data.plot(ax=ax2, kind="scatter", marker="o", s=45, color="red",
               y="Stress Level", x="Quality of Sleep", alpha=0.5,
               title="Female Stress Level Vs Quality of Sleep")

plt.tight_layout()
plt.show()
```

```python
male_data = sleep_health[sleep_health['Gender'] == 'Male']

# male sleep duration and quality of sleep
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#Male Sleep Duration vs Stress Level
male_data.plot(ax=ax1, kind="scatter", marker="o", s=45, color="blue",
               y="Stress Level", x="Sleep Duration", alpha=0.5,
               title="Male Stress Level Vs Sleep Duration")

#Male Quality of Sleep vs Stress Level
male_data.plot(ax=ax2, kind="scatter", marker="o", s=45, color="orange",
               y="Stress Level", x="Quality of Sleep", alpha=0.5,
               title="Male Stress Level Vs Quality of Sleep")

plt.tight_layout()
plt.show()
```

```python
female_data = sleep_health[sleep_health['Gender'] == 'Female']
male_data = sleep_health[sleep_health['Gender'] == 'Male']

# Create subplots for female and male stress levels vs sleep duration
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Scatter plot for Female Stress Levels vs Sleep Duration
female_data.plot(ax=ax1, kind="scatter", marker="o", s=45, color="pink", 
                 x="Sleep Duration", y="Stress Level", alpha=0.5,
                 title="Female Sleep Duration Vs Stress Levels")

# Scatter plot for Male Stress Levels vs Sleep Duration
male_data.plot(ax=ax2, kind="scatter", marker="o", s=45, color="blue", 
               x="Sleep Duration", y="Stress Level", alpha=0.5,
               title="Male Sleep Duration Vs Stress Levels")

plt.tight_layout()
plt.show()
```

```python
# average stress level, sleep duration, and quality of sleep for each gender
average_stress_level = sleep_health.groupby('Gender')['Stress Level'].mean()
average_sleep_duration = sleep_health.groupby('Gender')['Sleep Duration'].mean()
average_quality_of_sleep = sleep_health.groupby('Gender')['Quality of Sleep'].mean()
print(average_quality_of_sleep,average_sleep_duration,average_stress_level)
#subplots for each bar graph
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 12))

#bar graph of the average stress level for each gender
average_stress_level.plot(ax=ax1, kind='bar', color=['pink', 'blue'], alpha=0.7)
ax1.set_title('Average Stress Level by Gender')
ax1.set_ylabel('Average Stress Level')
ax1.set_xlabel('Gender')
ax1.set_xticks(range(len(average_stress_level)))
ax1.set_xticklabels(average_stress_level.index, rotation=0) 

#bar graph of the average sleep duration for each gender
average_sleep_duration.plot(ax=ax2, kind='bar', color=['pink', 'blue'], alpha=0.7)
ax2.set_title('Average Sleep Duration by Gender')
ax2.set_ylabel('Average Sleep Duration')
ax2.set_xlabel('Gender')
ax2.set_xticks(range(len(average_sleep_duration)))
ax2.set_xticklabels(average_sleep_duration.index, rotation=0)

# bar graph of the average quality of sleep for each gender
average_quality_of_sleep.plot(ax=ax3, kind='bar', color=['pink', 'blue'], alpha=0.7)
ax3.set_title('Average Quality of Sleep by Gender')
ax3.set_ylabel('Average Quality of Sleep')
ax3.set_xlabel('Gender')
ax3.set_xticks(range(len(average_quality_of_sleep)))
ax3.set_xticklabels(average_quality_of_sleep.index, rotation=0)

plt.tight_layout()
plt.show()
```

<b>How Sleep Duration Affects Quality of Sleep and Stress</b>

From the data represented in the above graphs, we can draw several conclusions. Primarily, longer sleep duration correlates with both better sleep quality and lower stress levels. When comparing genders, it appears that females, on average, experience better sleep outcomes than males. Specifically, they tend to sleep longer, report higher quality sleep, and have lower stress levels. Interestingly, for females, these benefits become more pronounced when sleep duration exceeds 7.5 hours
